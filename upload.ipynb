```python
import pandas as pd
import firebase_admin
from firebase_admin import credentials, firestore
```

```python
df = pd.read_csv("orginal_sales_data_edit.csv")
```

```python
df
```

```python
cred = credentials.Certificate("Private key from Firebase Admin SDK")
```

```python
app = firebase_admin.initialize_app(cred)
```

```python
dbFireStore = firestore.client()
```

```python
for item in df.to_dict(orient='record'):
    doc_ref = dbFireStore.collection('collection).document()
    doc_ref.set(item)
```

```python
queryResults = list(dbFireStore.collection(u'collection').stream())
listQueryResults = list(map(lambda x : x.to_dict(), queryResults))
df = pd.DataFrame(listQueryResults)
```
